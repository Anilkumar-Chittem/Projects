```python
# since there are outliers we need to remove them 
from scipy.stats import zscore
z_score=abs(zscore(data))
print(data.shape)
data_new=data.loc[(z_score<3).all(axis=1)]
print(data_new.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
run=sc.fit_transform(covid)
final=pd.DataFrame(run,columns=covid.columns)
```

```python
final.head()
```

```python
x=final.drop(columns='Deaths',axis=1)
y=final[['Deaths']]
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lg=LinearRegression()
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
def maxr2_score(lg,x,y):
    max_r_score=0
    for r_state in range(0,101):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.20)
        lg.fit(x_train,y_train)
        y_pred=lg.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print(" r2 score coressponding o ",r_state,'is',r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
        print("max r2 score coresspondnig to ",final_r_state,'is',max_r_score)
        return final_r_state
        
```

```python
from sklearn.model_selection import cross_val_score
r_state=maxr2_score(lg,x,y)
print(" Mean r2 score for linear regression is ",cross_val_score(lg,x,y,cv=5,scoring='r2').mean())
lg.fit(x_train,y_train)
print(lg.score(x_train,y_train))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor

from sklearn.neighbors import KNeighborsRegressor
neighbors={'n_neighbors':range(1,30)}
r=KNeighborsRegressor()
gknr=GridSearchCV(r,neighbors,cv=10)
gknr.fit(x,y)
gknr.best_params_
```

```python
knr=KNeighborsRegressor(n_neighbors=3)
knr.fit(x_train,y_train)
pred=knr.predict(x_test)
r2=r2_score(y_test,pred)
print(" r2 score : ",r2)
```
